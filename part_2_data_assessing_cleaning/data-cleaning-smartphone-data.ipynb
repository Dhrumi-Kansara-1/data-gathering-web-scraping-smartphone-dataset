#### solution

- store in no_of_cores -> if str contains 'core' string
- store in processor_speed -> if str contains 'Processor' substring
- remaing is name

```python
#define

#store in no_of_cores -> if str contains 'core' string
#store in processor_speed -> if str contains 'Processor' substring
#remaing in name
def divide(x):
  core=''
  speed=''
  name=''
  for i in x:
    if 'Core' in i:
      core=i
    elif 'GHz' in i:
      speed=i
    else:
      name=i
  return {'name':name, 'core':core, 'speed':speed}
```

```python
#define

#code
def num_of_cam(camera):
  if 'Quad' in camera:
    return 4
  elif 'Triple' in camera:
    return 3
  elif 'Dual' in camera:
    return 2
  else:
    return 1
```

```python
def extract_camera(list):
  primary_rear=''
  primary_front=''
  num_front=0
  num_rare=0
  for i in list:
    # print(i)
    if 'Rear' in i:
      primary_rear=i.split('+')[0].strip().split('\u2009MP')[0].strip()
      num_rear=num_of_cam(i)
    else:
      primary_front=i.split('+')[0].strip().split('\u2009MP')[0].strip()
      num_front=num_of_cam(i)
  return {'primary_front':primary_front, 'num_front':num_front, 'primary_rear':primary_rear, 'num_rear':num_rear}
```
